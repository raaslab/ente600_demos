# Simple Feature Matching Demo

This version is intentionally minimal. The notebook shows only the **main ideas**:

1. Read two images
2. Detect Harris corners
3. Extract simple patch descriptors
4. Match the descriptors
5. Visualize the matches

Most of the helper code is hidden in `feature_matching_utils.py`.

If you are running this in **Google Colab**, run the setup cell first.

```python
# Run this cell first in Google Colab.

REPO_URL = "https://github.com/raaslab/ente600_demos.git"
REPO_DIR = "ente600_demos"

def setup_demo(repo_url: str, repo_dir: str) -> None:
    import os
    import subprocess
    from pathlib import Path

    if Path("feature_matching_utils.py").exists():
        print("Already in the project folder.")
        return

    if not Path(repo_dir).exists():
        subprocess.run(["git", "clone", repo_url], check=True)

    os.chdir(repo_dir)
    print("Changed working directory to:", os.getcwd())


setup_demo(REPO_URL, REPO_DIR)
```

## Descriptor used here

We use a **normalized intensity patch** around each Harris corner.

That means:
- take a small image patch around each corner
- flatten it into a vector
- subtract the mean and divide by the standard deviation

This is still simple, but more robust than using a raw patch directly.

```python
from feature_matching_utils import (
    detect_harris_corners,
    extract_patch_descriptors,
    load_demo_images,
    match_descriptors,
    show_detected_corners,
    show_feature_matches,
    show_input_images,
    show_patch_descriptor_examples,
)
```

```python
image1, image2 = load_demo_images()

show_input_images(image1, image2)
```

```python
corners1 = detect_harris_corners(image1)
corners2 = detect_harris_corners(image2)

print('Corners in image 1:', len(corners1))
print('Corners in image 2:', len(corners2))

show_detected_corners(image1, corners1, image2, corners2)
```

```python
patch_size = 11

corners1, desc1 = extract_patch_descriptors(image1, corners1, patch_size=patch_size, normalize=True)
corners2, desc2 = extract_patch_descriptors(image2, corners2, patch_size=patch_size, normalize=True)

print('Descriptor type: normalized patch')
print('Patch size:', patch_size, 'x', patch_size)
print('Descriptor dimension:', desc1.shape[1])

show_patch_descriptor_examples(image1, corners1, desc1, patch_size=patch_size, num_examples=3)

matches = match_descriptors(desc1, desc2, ratio_thresh=0.75)
print('Number of matches:', len(matches))

show_feature_matches(image1, image2, corners1, corners2, matches, max_matches=25)
```
